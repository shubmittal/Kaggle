```python
#Option1 : Drop all columns with nulls
option1_dropna = sf_permits.dropna(axis =1)  
# this leaves 12 columns which is expected as we had total of 43 columns and 31 had one or more null values. 
# Axis =0 would have meant that drop any rows which have any nulls.

option1_dropna.shape
```

```python
sf_permits.shape
```

```python
#Option2 : Fill all columns with null with 0

option2_fill0 = sf_permits.fillna(0);

# Now check does option2 leave any nulls

nullcount = option2_fill0.isnull().sum();

nullcount
```

```python
#Option3: Fill all null values with values in the row before them

option3_fillwithvaluebefore = sf_permits.fillna(method = "bfill", axis=0)

option3_fillwithvaluebefore.isnull().sum()

#at this point you notice that certain values are still left as null. The reason is that there may not be any values before them to fill.
# So add fillna(0) to fill such values as well.

option3_fillwithvaluebefore = sf_permits.fillna(method = "bfill", axis=0).fillna(0)

option3_fillwithvaluebefore.isnull().sum()
```
